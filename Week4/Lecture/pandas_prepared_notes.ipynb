# pandas



*   **pandas** is a popular open-source **data analysis and manipulation library** for Python. It provides **data structures and functions needed** to work with structured data efficiently.


*   pandas is often **used in tandem with** numerical computing tools like **NumPy** and **SciPy**, analytical libraries like **statsmodels** and **scikit-learn**, and data visualization libraries like **matplotlib**.


*   pandas **incorporates many elements of NumPy's approach to array-based computing**, particularly its use of array-based functions and its emphasis on data processing **without the need for for loops**.


*   While pandas borrows many coding practices from NumPy, **the primary distinction** lies in their focus: **pandas is specifically designed for handling tabular or heterogeneous data, whereas NumPy is optimized for working with numerical arrays that have a single data type**.




# Key Features of pandas


*   **Data Cleaning:** Handle missing data, duplicate data, and data conversion.


*   **Data Transformation:** Perform operations like sorting, filtering, merging, and grouping data.


*   **Statistical Analysis:** Apply various statistical functions and aggregations.


*   **Input/Output Support:** Easily read from and write to different file formats like CSV, Excel, SQL databases, and more.


*   **High Performance:** Built on top of NumPy, which allows for fast data manipulation and computations.

# DataFrame



*   A **DataFrame** is **a rectangular data table** with an ordered, named collection of columns, where each column may contain a different data type (such as numeric, string, or Boolean).


*   It includes both row and column indices and can be viewed as a dictionary of Series that all share the same index.




There are several ways to create a DataFrame, but one of the most common methods is using a dictionary of equal-length lists or NumPy arrays.

```python
d = {
    "Name": ["John", "Alice", "Jack", "Rose", "Alex", "David"],
    "Age": [32, 23, 21, 35, 45, 48],
    "City": ["Istanbul", "Paris", "London", "Madrid", "Sao Paulo", "Miami"]
}

data = pd.DataFrame(d)
data
```

# DataFrame - deleting a column/row

We can use **`drop()`** method with **`axis`** parameter.

*   When **`axis`** is 0, it means deleting from the 0th dimension (row).
*   When the **`axis`** is 1, the it means deleting from 1st dimension (column).
*   By default, **drop()** method does not work inplace, and returns a new dataframe. So the original DataFrame will not be modified.




```python
data
```

```python
# returns another dataframe
data.drop("debt", axis=1)
```

```python
# but data is still same
data
```

If you want to apply this delete operation on original DataFrame, then there are two different ways:


*   `data = data.drop("debt", axis=1) # or`
*   `data.drop("debt", axis=1, inplace=True)`




```python
data.drop("debt", axis=1, inplace=True)
data
```

In order to delete a row, set **`axis`** parameter as 0.

```python
data.drop(2, axis=0, inplace=True)
data
```

```python
data.drop([1,5], axis=0, inplace=True)
data.drop(["Age","City"], axis=1, inplace=True)
data
```

# and (&) / or (|) oeprators

`&` -> and

`|` -> or

```python
movies[(movies["duration"] > 200) & (movies["genre"] == "Drama")]
```

The following will not work, since pandas expects you to be explicit with expressions by using parentheses.

```python
movies[movies["duration"] > 200 & movies["genre"] == "Drama"]
```

If there are multiple expressions, then you can execute them on separate lines without using parentheses.

```python
c1 = movies["duration"] > 200
c2 = movies["genre"] == "Drama"
movies[c1 & c2]
```

Let's write a more complex expression

```python
movies[(movies["duration"] > 200) & ((movies["genre"] == "Drama") | (movies["genre"]=="Crime"))]
```

Another way of writing previous expression with **`isin()`** function

```python
movies[(movies["duration"] > 200) & (movies["genre"].isin(["Drama", "Crime"]))]
```

# Type Conversion

We can apply type conversion with **`astype()`** method.

```python
orders = pd.read_table("http://bit.ly/chiporders")
orders.head()
```

The following returns a Series with **`object`** dtype (because of the $ sign). So we cannot calculate the mean of the values directly.

```python
orders["item_price"]
```

Let's remove $ sign with **`replace()`** method.

In pandas, the `str` accessor is used when applying string functions to the elements of a Series. This is necessary because string operations are not directly available on the Series object, which might contain different data types. Using the `str` accessor ensures that the operation is applied to each element of the Series as a string, even if the Series itself isn't explicitly of type str.

```python
orders["item_price"].str.replace("$", "")
```

or we can use **`slice()`** method to remove $ sign

```python
orders["item_price"].str.slice(1)
```

Notice that previous results are still retrived in object dtype. Let's convert it into float.

```python
orders["item_price"].str.slice(1).astype(float)
```

Now, ready to calculate the mean

```python
orders["item_price"].str.slice(1).astype(float).mean()
```

Let's calculate the mean price of the items whose name is "Chicken Bowl"

```python
orders[orders["item_name"] == "Chicken Bowl"]["item_price"].str.slice(1).astype(float).mean()
```

Let's calculate the mean price of items that have the word "chicken" in their name, regardless of its case, such as "chicken," "chiCKen," etc.

```python
Chicken_df = orders[orders["item_name"].str.lower().str.contains("chicken")]
Chicken_df["item_price"].str.slice(1).astype(float).mean()
```

# Grouping Operations

```python
drinks = pd.read_csv("http://bit.ly/drinksbycountry")
drinks.head()
```

Let's find average beer consumption in Europe countries

```python
drinks[drinks.continent == "Europe"]["beer_servings"].mean()
```

```python
drinks.continent.unique()
```

Now let's find average beer consumption for all continents

```python
for cont in drinks.continent.unique():
  res = drinks[drinks.continent == cont]["beer_servings"].mean()
  print(cont, res)
```

But there is an easier way with **`groupby()`** method.

```python
# we need to use numeric_only parameter since there are also non-numeric columns
drinks.groupby("continent").mean(numeric_only=True)
```

```python
# for the following we don't need to use numeric_only parameter
drinks.drop("country", axis=1).groupby("continent").mean()
```

```python
drinks.groupby("continent").mean(numeric_only=True)["beer_servings"]
```

```python
drinks.groupby("continent")["beer_servings"].mean().index
```

```python
drinks.groupby("continent")["beer_servings"].mean().values
```

# Group sizes and **`value_count()`** method

There are 2 ways to find group sizes (like how many time we observe each continent in our dataset):


*   We can use **`frame.groupby(columnName).size()`** option. For instance **`drinks.groupby("continent").size()`** will find frequencies of each continent, but they will be displayed based on alphabetical (ascending) order.
*   We can use **`frame[columnName].value_counts()`** option. For instance **`drinks["continent"].value_counts()`** will find frequencies of each continent, but they will be displayed based on their frequency value (in descending order).



```python
drinks.groupby("continent").size()
```

```python
drinks["continent"].value_counts()
```

We can sort the last result based on index and get the same result with groupby() method

```python
drinks["continent"].value_counts().sort_index()
```

```python
drinks.groupby("continent").count()
```

We can use multiple aggregations, however we need to drop country column to execute following line

```python
drinks.groupby("continent").agg(["max", "mean","min"])
```

```python
drinks.drop("country", axis=1).groupby("continent").agg(["max", "mean","min"])
```

You can also set **`figsize`** parameter within the **`plot()`** method.

```python
drinks.drop("country", axis=1).groupby("continent").agg(["max", "mean","min"]).plot(kind="bar", figsize=(10, 6))
```

Print the result (actually any DataFrame) into a excel file with **`to_excel()`**

```python
drinks.drop("country", axis=1).groupby("continent").agg(["max", "mean","min"]).to_excel("results.xlsx")
```
