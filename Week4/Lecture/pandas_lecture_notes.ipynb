and -> &

or -> |

```python
# find the movies longer than 3 hours and genre is "Drama"

movies[(movies["duration"] > 180) & (movies["genre"] == "Drama")]
```

```python
c1 = movies["duration"] > 180
c2 = movies["genre"] == "Drama"

movies[c1 & c2]
```

```python
# find the movies longer than 3 hours and genre is Drama or Adventure
movies[(movies["duration"] > 180) & ((movies["genre"] == "Drama") | (movies["genre"] == "Adventure"))]
```

```python
#movies[(movies["duration"] > 180) & ((movies["genre"] == "Drama") | (movies["genre"] == "Adventure"))]
movies[(movies["duration"] > 180) & (movies["genre"].isin(["Drama", "Adventure"]))]
```

```python
orders = pd.read_table("http://bit.ly/chiporders")
orders.head()
```

```python
# find the average price for each record

orders["item_price"].str.replace("$", "").astype(float).mean()
```

```python
orders["item_price"].str.slice(1).astype(float).mean()
```

```python
# calculate avg price for "Chicken Bowl" items

orders[orders["item_name"] == "Chicken Bowl"]["item_price"].str.slice(1).astype(float).mean()
```

```python
df = orders[orders["item_name"] == "Chicken Bowl"]
df["item_price"].str.slice(1).astype(float).mean()
```

```python
# calculate the avg price for the items whose name contains "chicken" or "ChiCken" or etc

temp = orders[orders["item_name"].str.lower().str.contains("chicken")]
temp["item_price"].str.slice(1).astype(float).mean()
```

```python
drinks = pd.read_csv("http://bit.ly/drinksbycountry")
drinks.head()
```

```python
# calculate the average wine_cons in Europe

drinks[drinks["continent"] == "Europe"]["wine_servings"].mean()
```

```python
drinks["continent"].unique()
```

```python
for cont in drinks["continent"].unique():
    res = drinks[drinks["continent"] == cont]["wine_servings"].mean()
    print(cont, res)
```

```python
drinks.groupby("continent")["wine_servings"].mean()
```

```python
drinks.groupby("continent").mean(numeric_only=True)
```

```python
drinks.drop("country", axis=1).groupby("continent").mean()
```

```python
drinks.groupby("continent").min()
```

```python
drinks.groupby("continent").max()
```

```python
drinks[drinks["continent"] == "Europe"].sort_values("beer_servings", ascending=False)
```

```python
drinks.groupby("continent").mean(numeric_only=True)["beer_servings"].plot(kind="bar")
```

```python
temp = drinks.groupby("continent").mean(numeric_only=True)["beer_servings"]
plt.bar(temp.index, temp.values)
plt.xticks(rotation=90)
plt.show()
```

```python
drinks.groupby("continent").mean(numeric_only=True).plot(kind="bar", figsize=(10, 6))
```

```python
temp = drinks.groupby("continent").mean(numeric_only=True)
temp.to_csv("result.csv")
```
