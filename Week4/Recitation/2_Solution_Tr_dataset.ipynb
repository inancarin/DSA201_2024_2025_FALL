```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

Read the `tr.txt`

```python
data = pd.read_csv("tr.txt")
data.head()
```

Check the shape of the data

```python
data.shape
```

Check the data types of the columns (Series)

```python
data.dtypes
```

Plot the Türkiye map as a scatter plot

```python
plt.figure(figsize=(10, 5))
plt.scatter(data["lng"], data["lat"])
plt.show()
```

```python
# or
data.plot(x="lng", y="lat", kind="scatter", figsize=(10, 5))
```

Find the most crowded 20 places based on `population` column and plot them with a bar chart

```python
temp = data.sort_values("population", ascending=False)[:20]
plt.bar(temp["city"], temp["population"])
plt.xticks(rotation=90)
plt.xlabel("city")
plt.show()
```

```python
# or
data.sort_values("population", ascending=False)[:20].plot(kind="bar", x="city", y="population")
```

Find the top 20 provinces (admin_name) with the highest number of prominent cities and display the result with a bar chart

```python
data["admin_name"].value_counts()[:20].plot(kind="bar")
```

Find the name of the İstanbul's the most easterly location. You just need to print the result as a string.

```python
ist_data = data[data.admin_name == "İstanbul"]
maxLng = ist_data["lng"].max()
ist_data[ist_data["lng"] == maxLng]["city"].values[0]
```
