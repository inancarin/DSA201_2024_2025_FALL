```python
f = open("iris.txt")

allLines = f.readlines()

x_values = []
y_values = []
species_to_colors = {}
color_codes = []
for line in allLines[1:]:
    line = line.strip()
    info = line.split(",")

    sepalLength = float(info[1])
    sepalWidth = float(info[2])
    species = info[-1]

    if species not in species_to_colors:
        species_to_colors[species] = len(species_to_colors)

    color_codes.append(species_to_colors[species])
    x_values.append(sepalLength)
    y_values.append(sepalWidth)


f.close()

print(species_to_colors)
print(color_codes)
plt.scatter(x_values, y_values, c=color_codes)
plt.xlabel("SepalLength in cm")
plt.ylabel("SepalWidth in cm")
plt.show()
```

```python
f = open("iris.txt")

allLines = f.readlines()

d = {}

for line in allLines[1:]:
    line = line.strip()
    info = line.split(",")

    sepalLength = float(info[1])
    sepalWidth = float(info[2])
    species = info[-1]

    if species not in d:
        d[species] = [[sepalLength],[sepalWidth]]
    else:
        d[species][0].append(sepalLength)
        d[species][1].append(sepalWidth)
        
f.close()

for k, v in d.items():
    plt.scatter(v[0], v[1], label=k)

plt.legend()
plt.xlabel("SepalLength in cm")
plt.ylabel("SepalWidth in cm")
plt.show()
```

# Example 4

```python
f = open("GOOGL.txt")

allLines = f.readlines()

d = {}

for line in allLines[1:]:
    line = line.strip()
    info = line.split(",")
    date = info[0]
    year = int(date.split("-")[0])
    if 2004 < year < 2022:
        close = float(info[4])
        if year not in d:
            d[year] = {"days": [], "stocks": []}
        d[year]["days"].append(len(d[year]["days"])+1)
        d[year]["stocks"].append(close)

f.close()

plt.figure(figsize=(20, 10))
for k, v in d.items():
    plt.plot(v["days"], v["stocks"], label=k)

plt.xlabel("Day of the Year")
plt.ylabel("Stock Prices")
plt.legend()
plt.show()
```
