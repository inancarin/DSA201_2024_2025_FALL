```python
import matplotlib.pyplot as plt
import numpy as np
```

# Question 1

## 1st solution

```python
freqs = [0] * 11 # we could also use dictionary instead

for i in range(100000):
    num1 = np.random.randint(1, 7) # dice1
    num2 = np.random.randint(1, 7) # dice2
    total = num1 + num2
    idx = total - 2
    freqs[idx] += 1

plt.bar(range(2,13), freqs)
plt.xticks(range(2,13), rotation=60)
plt.xlabel("Sum Value")
plt.ylabel("Freqs")
plt.show()
```

## 2nd solution

```python
freqs = [0] * 11


mat1 = np.random.randint(1, 7, (1, 100000)) # dice1
mat2 = np.random.randint(1, 7, (1, 100000)) # dice2
mat = np.concatenate((mat1, mat2), axis=0)
sums = mat.sum(axis=0)

for val in sums:
    idx = val - 2
    freqs[idx] += 1

plt.bar(range(2,13), freqs)
plt.xticks(range(2,13), rotation=60)
plt.xlabel("Sum Value")
plt.ylabel("Freqs")
plt.show()
```

## 3rd solution

```python
d = {}

for i in range(100000):
    num1 = np.random.randint(1, 7) # dice1
    num2 = np.random.randint(1, 7) # dice2
    total = num1 + num2
    if total not in d:
        d[total] = 1
    else:
        d[total] += 1


plt.bar(d.keys(), d.values())
plt.xticks(range(2,13), rotation=60)
plt.xlabel("Sum Value")
plt.ylabel("Freqs")
plt.show()
```

# Question 2

```python
f = open("temps.txt", "r")

lines = f.readlines()

f.close()

d = {}

for line in lines[1:]:
    line = line.strip()
    info = line.split("\t")
    city = info[1]
    temp = int(info[2].replace("C", ""))

    if city not in d:
        d[city] = [temp]
    else:
        d[city].append(temp)


plt.plot(range(1,13) , d["Istanbul"])
plt.xticks(range(1,13))
plt.xlabel("Months")
plt.ylabel("Average Temp (C)")
plt.grid(True, alpha=0.3)
plt.show()
```

# Question 3

```python
tr_file = open("tr.txt")

lines = tr_file.readlines()

tr_file.close()

lats, lngs = [], []
for line in lines[1:]:
    line = line.strip()
    info = line.split(",")
    lat = float(info[1])
    lng = float(info[2])
    lats.append(lat)
    lngs.append(lng)

plt.figure(figsize=(10, 5))
plt.scatter(lngs , lats)
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.show()
```

## Solution with c parameter

```python
tr_file = open("tr.txt")

lines = tr_file.readlines()

tr_file.close()

lats, lngs, codes = [], [], []
d = {}
n = 0
for line in lines[1:]:
    line = line.strip()
    info = line.split(",")
    lat = float(info[1])
    lng = float(info[2])
    city = info[5]
    if city not in d:
        d[city] = n
        n += 1
    codes.append(d[city])
    lats.append(lat)
    lngs.append(lng)

plt.figure(figsize=(10, 5))
plt.scatter(lngs , lats, c=codes)
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.show()
```

## Solution without c parameter 

```python
tr_file = open("tr.txt")

lines = tr_file.readlines()

tr_file.close()

d = {}
for line in lines[1:]:
    line = line.strip()
    info = line.split(",")
    lat = float(info[1])
    lng = float(info[2])
    city = info[5]
    if city not in d:
        d[city] = {}
        d[city]["lat"] = [lat]
        d[city]["lng"] = [lng]
    else:
        d[city]["lat"].append(lat)
        d[city]["lng"].append(lng)

print(d["stanbul"]["lat"]) # I just wanted to check 
print(d["stanbul"]["lng"])

plt.figure(figsize=(10,5))
for k, v in d.items():
  plt.scatter(v["lng"], v["lat"])
plt.show()
```
