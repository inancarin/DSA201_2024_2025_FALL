```python
for i in range(1000):
    temp = titanic.sample(5, replace=False).index
    if len(temp.unique()) != 5:
        print(temp)
```

```python
for i in range(1000):
    temp = titanic.sample(5, replace=True).index
    if len(temp.unique()) != 5:
        print(temp)
```

```python
allAverages = []
for i in range(1000):
    temp = titanic["Fare"].sample(100, replace=True)
    cur_mean = temp.mean()
    allAverages.append(cur_mean)
```

```python
sns.displot(allAverages)
```

```python
x = 95
thr = (100-x) / 2
thr_index = int(thr * len(allAverages) / 100)
allAverages.sort()
thr1 = allAverages[thr_index]
thr2 = allAverages[-thr_index]
print(thr1, thr2)
```

```python
sns.displot(allAverages)
plt.plot([thr1, thr1], [0, 120], color="red", linestyle="dashed", linewidth=3)
plt.plot([thr2, thr2], [0, 120], color="red", linestyle="dashed", linewidth=3)
plt.show()
```
