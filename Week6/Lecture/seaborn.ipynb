# barplot()

Seaborn's **`barplot()`** is a function used to create bar plots, where the height of the bars represents a summary statistic (such as the mean, median, or sum) of a numerical variable for each category of a categorical variable. By default, **`barplot()`** shows the mean value of the numerical variable for each category and includes error bars representing confidence intervals.

*   **Categorical vs. Numerical Data:** It expects a categorical variable on the x-axis (or y-axis, depending on the orientation) and a numerical variable on the other axis.
*   **Summary Statistic:** By default, the summary statistic is the mean, but you can specify a different statistic (e.g., median, sum) using the estimator parameter.
*   **Error Bars:** It adds error bars to represent uncertainty or variability (95% confidence intervals by default).

We can visualize the average bill paid with respect to the day of the week using a bar chart in Seaborn. We can achieve this by calculating the mean of the **`total_bill`** for each day and then plotting it using Seaborn's **`barplot`**() function.

```python
#Same thing with seaborn
sns.barplot(x="day", y="total_bill", data=tips)
#sns.barplot(x="day", y="total_bill", data=tips, estimator="mean")
```

We can create a bar plot in Seaborn without using the **`data`** parameter by directly passing the values for the **`x`** and **`y`** parameters as separate lists, pandas Series, or arrays. This approach allows you to work without using the dataset keyword argument (data). However, we will need to manually extract the data from the dataset and pass it as arguments to **`barplot()`**.

Here's how you can resolve the previous example without using the **`data`** parameter:

```python
sns.barplot(x=tips["day"], y=tips["total_bill"])
```

The **`hue`** parameter in Seaborn is used to add a third categorical variable (in addition to **`x`** and **`y`**) to a plot, which groups and colors the data points based on the values of this third variable. It essentially enables "conditional" grouping, where different subsets of data are displayed in different colors to highlight distinctions between groups. This is particularly useful when you want to visualize differences between subgroups within your main categorical variable.

```python
#We can even divide each day into 2 categories with respect to sex:
sns.barplot(x="day", y="total_bill", data=tips, hue="sex")
```

The **`palette`** parameter in Seaborn is used to control the color scheme of a plot. It allows you to specify the colors that will be used to represent different categories in categorical plots or to distinguish different elements in visualizations like bar charts, scatter plots, or line plots.

Here are the most common types of palettes (do not try to memorize them, you can always check them from the documentation):

1. **Qualitative Palettes (for categorical data):**'deep' (default), 'muted', 'bright', 'pastel', 'dark', 'colorblind'

2. **`Sequential Palettes (for ordered or continuous data):`** 'Blues', 'Greens', 'Reds', 'Purples'

3. **`Diverging Palettes (for data with a midpoint, often centered at zero):`** 'RdBu', 'PuOr', 'BrBG', 'Spectral'


```python
#Change the palette business
sns.barplot(x="day", y="total_bill", data=tips, hue="sex", palette="autumn")
```

The **`estimator`** parameter in Seaborn is used in functions like **`barplot()`**, to specify which statistical function should be applied to the data for aggregation. By default, the mean is used, but you can change the estimator to any other function, such as median, sum, or a custom function.

n Seaborn's **`barplot()`**, **`error bars`** are the vertical lines that extend above and below the top of each bar, representing the uncertainty or variability in the data. They provide a visual indication of how precise the estimate of the central tendency (such as the mean) is. These error bars are typically derived from confidence intervals or standard deviation, depending on the parameter settings in the **`barplot()`** function.

By default, the error bars represent 95% confidence intervals around the estimate (such as the mean). This indicates that there is a 95% probability that the true value (e.g., the true population mean) lies within the range of the error bars. In order to disable error bars, you can set **`errorbar=None`**.

```python
sns.barplot(x="day", y="total_bill", data=tips, hue="sex", palette="autumn", estimator=np.median, errorbar=None)
```

# violinplot()


A violin plot in Seaborn is a method of visualizing the distribution of data across categories. It is similar to a boxplot but with additional information about the probability density of the data at different values. It combines aspects of both boxplots and kernel density plots, making it a useful tool for visualizing the spread, distribution, and shape of the data.

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

The **`cut`** parameter in Seaborn's violinplot() controls how far the kernel density estimate (KDE) extends beyond the extreme data points (i.e., beyond the minimum and maximum values). By default, Seaborn extends the KDE beyond the actual range of the data to provide a smoother and more informative density curve. However, with the **`cut`** parameter, you can adjust or limit this extension.
- Higher cut values: Extend the KDE further beyond the data range, making the plot appear smoother.
- Lower cut values (including cut=0): Restrict the KDE to the actual range of the data, showing only the density within the minimum and maximum of the dataset.

```python
sns.violinplot(x="day", y="total_bill", data=tips, cut=0)
```

```python
plt.figure(figsize = (10,7))
sns.violinplot(x="day", y="total_bill", data=tips, cut=0, hue="sex")
```

With **`split=True`**, the two distributions (one for each category in hue) are split into two halves of the same violin, with each half showing one category's density. This provides a more compact and direct comparison between two categories within the same plot.

```python
plt.figure(figsize = (10,7))
sns.violinplot(x="day", y="total_bill", data=tips, cut=0, hue="sex", split=True)
```

# Some Statistical Concepts

## Sampling with Pandas

In statistics, sampling is the process of selecting a subset from a larger population to estimate characteristics of the whole population. In this question, we will talk about two sampling methods: Sampling With Replacement and Sampling Without Replacement.

**Sampling with Replacement:** Let's say you had a population of 7 people, and you wanted to sample 2. Their names are: John, Jack, Qiu, Tina, Hatty, Jacques, Des. You could put their names in a hat. If you sample with replacement, you would choose one person's name, put that person's name back in the hat, and then choose another name. The possibilities for your two-name sample are: (John, John), (John, Jack), (John, Qui), (Jack, Qui) etc. There are exactly 7*7=49 different possibilities.

**Sampling Without Replacement:** Sampling without Replacement is a way to figure out probability without replacement. In other words, you don't replace the first item you choose before you choose a second. Taking the above example, you would have the same list of names to choose two people from. And your list of results would be similar, except you couldn't choose the same person twice: (John, Jack), (John, Qui), (Jack, Qui) etc. Now we have 7*6=42 different possibilities here.

In pandas, we have a method called **`sample()`** which can be used for the sampling process.

**`data.sample(5, replace=True)`** `# randomly select 5 samples with replacement`

**`data.sample(5, replace=False)`** `# randomly select 5 samples without replacement`

**`data[columnName].sample(5, replace=True)`** `# randomly select 5 samples from columnName with replacement`

**`data[columnName].sample(5, replace=False)`** `# randomly select 5 samples from columnName without replacement`

Let's use the [titanic data](https://bit.ly/kaggletrain) that we used in the lecture.

Firstly, select random 5 samples without replacement. Then repeat this 1000 times, and show that all of the samples within each iteration are unique.

```python
df = pd.read_csv("http://bit.ly/kaggletrain")

for i in range(1000):
  temp = df.sample(5, replace=False)
  indices = temp.index
  if len(indices.unique()) != 5:
    print(indices)

# you will see no output, since all of them are unique numbers
```

Now, select random 5 samples with replacement. Then repeat this 1000 times, and show that some of the samples within each iteration may not be unique.

```python
for i in range(1000):
  temp = df.sample(5, replace=True)
  indices = temp.index
  if len(indices.unique()) != 5:
    print(indices)

# now you will probably see some output for some of the iterations
```

## Confidence Intervals

One way to think of a 90% confidence interval is as follows: it is the interval that encloses the central 90% of the bootstrap sampling distribution of a sample statistic. More generally, an x% confidence interval around a sample estimate should, on average, contain similar sample estimates x% of the time.

Given a sample of size n (let's say 100), and a sample statistic of interest, the algorithm for a bootstrap confidence interval is as follows:



1.   Draw a random sample of size n (100) with replacement from the data.
2.   Record the statistic of interest (say calculate and record mean) for the resample.
3.   Repeat steps 1 and 2 many (let's say 1000) times.
4.   For an x% confidence interval, trim `[(100-x) / 2]`% of the 1000 resample results from either end of the distribution.
5.   The trim points are the endpoints of an x% bootstrap confidence interval.

Plot a 95% confidence interval for the mean ticket fares in titanic, based on a sample of 100 as below (you may set **`linestyle="dashed"`** parameter of **`plot()`** function in matplotlib to draw a dashed line):

download.png



```python
# surely the result can be slightly different since it contains some randomness

n_iter = 1000
results = []
for i in range(n_iter):
  temp = df["Fare"].sample(100, replace=True)
  results.append(temp.mean())

results.sort()

x = 95
thr = (100-x)/2
print("thr:", thr)

thr_count = int((n_iter * thr) / 100)
print("thr_count:", thr_count)

left_line = results[thr_count]
right_line = results[-thr_count]
print("left:", left_line, ", right:", right_line)

sns.displot(results)
plt.plot([left_line, left_line], [0, 130], color="red", linestyle="dashed", label=format(left_line, ".2f"))
plt.plot([right_line, right_line], [0, 130], color="red", linestyle="dashed", label=format(right_line, ".2f"))
plt.legend()
plt.show()
```
